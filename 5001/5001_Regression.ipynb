```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('housing.csv') 
#Download from https://www.kaggle.com/datasets/camnugent/california-housing-prices?select=housing.csv
```

```python
data.dropna(inplace=True)
data = data.join(pd.get_dummies(data.ocean_proximity)).drop(['ocean_proximity'], axis=1)
#(GeeksforGeeks, 2020)
```

```python
standard_distribution = data.copy()
standard_distribution["total_rooms"] = np.log(data["total_rooms"])
standard_distribution["total_bedrooms"] = np.log(data["total_bedrooms"])
standard_distribution["population"] = np.log(data["population"])
standard_distribution["households"] = np.log(data["households"])
#   (Ryan & Matt Data Science, 2023)
```

```python
standard_distribution.hist(figsize=(20,8))
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(data.corr(), annot=True, cmap="RdYlGn")
```

```python
plt.figure(figsize=(15,8))
sns.scatterplot(x='longitude', y='latitude', hue='median_house_value', data=data, palette='RdYlGn_r')
```

```python
x = data.drop('median_house_value', axis=1)
y = data['median_house_value']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
train_data = x_train.join(y_train)
```

```python
#Linear Regression

regressor = LinearRegression()
x_train, y_train = train_data.drop('median_house_value', axis=1), train_data['median_house_value']
regressor.fit(x_train, y_train)

print(f"Linear Regression Train R2 = {regressor.score(x_train, y_train)}")
print(f"Linear Regression Test R2 = {regressor.score(x_test, y_test)}")
```

```python
#Random Forest Regressor
from sklearn.ensemble import RandomForestRegressor

forest = RandomForestRegressor()
forest.fit(x_train, y_train)

print(f"Random Forest Train R2 = {forest.score(x_train, y_train)}")
print(f"Random Forest Test R2 = {forest.score(x_test, y_test)}")
```

```python
#Forest Scatter Graph

forest_pred = regressor.predict(x_test)
plt.scatter(y_test, forest_pred, color='purple', label='Predicted', alpha=0.1)
plt.plot(y_test, y_test, color='red', label='Actual', linewidth=2, alpha=0.7)
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Median House Price (Tree) | Actual vs Predicted")
```

```python
#Linear Scatter Graph

linear_pred = regressor.predict(x_test)
plt.scatter(y_test, linear_pred, color='blue', label='Predicted', alpha=0.1)
plt.plot(y_test, y_test, color='red', label='Actual', linewidth=2, alpha=0.7)
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Median House Price (Linear) | Actual vs Predicted")
#(GeeksforGeeks, 2024)
```

```python
#Linear MAPE

from sklearn.metrics import mean_absolute_percentage_error

linear_MAPE = mean_absolute_percentage_error(y_test, linear_pred)
print(f"Linear - Mean Absolute Percentage Error (MAPE): {linear_MAPE:.2%}")

#Forest MAPE

forest_MAPE = mean_absolute_percentage_error(y_test, forest_pred)
print(f"Forest - Mean Absolute Percentage Error (MAPE): {forest_MAPE:.2%}")

#(scikit-learn, n.d.)
```
